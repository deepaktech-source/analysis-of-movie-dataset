# Data cleaning and transforming

SPLIT

```python
movies['Genres']=movies['Genres'].str.split(',')
```

```python
movies
```

```python
# movies_no_missing = movies.dropna(subset=['Rotten Tomatoes'])
```

```python
movies.isnull().values.any()
```

```python
movies.isnull().sum()
```

used astype() to change the data type of object to float

```python
movies['Rotten Tomatoes'] = movies['Rotten Tomatoes'].str.replace('%', '').astype(float)
```

In the nan values of rotten tomatoes using fillna() we filled zero instead of null values

```python
movies['Rotten Tomatoes'] = movies['Rotten Tomatoes'].fillna(0)
```

to find how many total null value

```python
movies.isnull().sum().sum()
```

we have some nan values in language column instead of that we fill with 'unknown'


```python
movies['Language']=movies['Language'].fillna('Unknown' )
```

```python
movies['Directors']=movies['Directors'].fillna('Unknown' )
```

```python
movies['Country']=movies['Country'].fillna('Unknown' )
```

```python
movies['Genres']=movies['Genres'].fillna('Unknown')
```

filling nan value of runtime with mean

```python
movies['Runtime']=movies['Runtime'].fillna(movies['Runtime'].mean() )
```

```python
# movies['IMDB']=movies['IMDB'].fillna('0' )
```

AGE

```python
#  Identify rows where 'Age' is 'all'
all_age_rows = movies[movies['Age'] == 'all'].index

# Replace 'all' values with 0 in 'Age' column
movies.loc[all_age_rows, 'Age'] = 0

movies['Age'] = movies['Age'].str.replace('+','')
movies['Age']= movies['Age'].fillna(0)
movies['Age'] = movies['Age'].astype(int)
movies['Age']=movies['Age'].fillna(movies['Age'].mode() )
```

To drop the unnamed and type column in dataset

```python
movies.drop(['Unnamed: 0','Type'],axis=1,inplace=True)
```

rename the column came as IMDB in IMDb

```python
movies.rename(columns={'IMDb':'IMDB'},inplace=True)
movies['IMDB']=movies['IMDB'].fillna('0')
movies['IMDB']=movies['IMDB'].astype(int)
```

8.Count of Movies by Age Rating

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=movies, x='Age',color='red',hue='Age',palette='colorblind')
plt.title('Count of Movies by Age Rating')
plt.xlabel('Age Rating')
plt.ylabel('Number of Movies')
plt.legend()
plt.show()
```

9.KDE plot for IMDB ratings(average)[KDE-Kernal Density Estimation]

```python
sns.kdeplot(data=movies, x='IMDB',color='skyblue',fill=True)
plt.title('KDE Plot of IMDB Ratings')
plt.xlabel('IMDB Rating')
plt.ylabel('Average')
plt.show()
plt.ylabel('Average')
plt.show()
plt.ylabel('Average')
plt.show()
plt.ylabel('Average')
plt.show()
```

10.number of movies released based on the year

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=movies, x='Year')
plt.title('movie released year')
plt.xlabel('year')
plt.ylabel('NO. OF movies')
plt.show()
```

11.Age's of the movies viewer

```python
sns.boxplot(data=movies, x='IMDB', color='brown')
plt.title('Average IMDB rate in movies')
plt.xlabel('IMDB Rating')
plt.grid()
plt.show()
plt.show()
```

12.Number of Movies on Each Streaming Platform

```python
platform_counts = data[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum()

plt.figure(figsize=(8, 6))
sns.barplot(x=platform_counts.index, y=platform_counts.values,color='red')
plt.title('Number of Movies on Each Streaming Platform')
plt.xlabel('Streaming Platform')
plt.ylabel('Number of Movies')
plt.show()

most_movies_platform = platform_counts.idxmax()
print(f"The platform with the most movies is: {most_movies_platform}")
```

13.languages are most commonly associated with high IMDB ratings

```python
language_imdb = movies.groupby('Language')['IMDB'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
print(language_imdb)

top_languages_by_rating = language_imdb.nlargest(10,'mean')
plt.figure(figsize=(10, 6))

sns.barplot(x=top_languages_by_rating.index, y='mean', data=top_languages_by_rating,)
plt.title('Top 10 Languages by Average IMDB Rating')
plt.xlabel('Language')
plt.ylabel('Average IMDB Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

14.Scatter plot ofIMDB and Rotten Tomatoes

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=movies, x='IMDB', y='Rotten Tomatoes',hue='IMDB',palette='colorblind')
plt.title('IMDB and Rotten Tomatoes')
plt.xlabel('IMDB Rating')
plt.ylabel('Rotten Tomatoes Rating')
plt.show()
```

15.Jointplot of 'Netflix and Runtime

```python
plt.figure(figsize=(5, 5))
sns.jointplot(data=movies, x='Netflix', y='Runtime', kind='reg')
plt.title('Netflix and Runtime')
plt.xlabel('Netflix')
plt.yticks([30,60,90,120,150,180])
plt.ylabel('Runtime')
plt.show()
plt.tight_layout()
plt.show()
```

16.lmplot IMDB and Rotten Tomatoes by Age

```python
sns.lmplot(data=movies, x='IMDB', y='Rotten Tomatoes', hue='Age')
plt.title('IMDB and Rotten Tomatoes by Age')
plt.xlabel('IMDB Rating')
plt.ylabel('Rotten Tomatoes Rating')
plt.show()
```

17.Pairwise Relationships between IMDB, Rotten Tomatoes, and Runtime

```python
sns.pairplot(data=movies, vars=['IMDB', 'Rotten Tomatoes', 'Runtime'])
plt.suptitle('Pairwise Relationships between IMDB, Rotten Tomatoes, and Runtime', y=1.02)
plt.show()
```

18.Is there a relationship between the age rating and other features like IMDB score and runtime using heatmap

```python
age_corr = movies.groupby('Age')[['IMDB', 'Rotten Tomatoes','Runtime']].mean()
print(age_corr)

plt.figure(figsize=(12,6))
sns.heatmap(age_corr.corr(), annot=True)
plt.show()
```

19.IMDB Distribution for Movies with more than 7

```python
high_rated_movies = movies[movies['IMDB'] >= 7]
plt.figure(figsize=(8, 6))
sns.histplot(data=high_rated_movies, x='IMDB', bins=20, kde=True, color='purple',hue='IMDB', palette= 'viridis')
plt.title('IMDB Distribution for Movies with more than 7')
plt.xlabel('IMDB Rating')
plt.ylabel('movies')
plt.show()
```
